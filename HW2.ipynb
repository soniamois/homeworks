```python
#создаем переменную
text= "варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг"""

#с помощью команды split считаем кол-во слов и выодим 
word= text.split()
bigrams = []
print(word)

#задаем цикл по созданию списка биграмм
for i in range(len(word)-1):
    bigrams.append([word[i],word[i+1]])
print(bigrams)
```

```python
text= """Несчастная кошка порезала лапу-
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!"""

letters = list(text)  # разбивает строку на минимальные элементы, т.е. посимвольно
print(letters)
print(len(text)) # №2


# выводим слова без знаков препинания
from string import punctuation

words = text.split()
words = [w.strip(punctuation) for w in words]
print(words) # №3

new_string = ""
for w in words:
    new_string += (w + " ")
print (new_string) # №4

print ([word.lower() if w.istitle() else word for word in words]) # №5

print (len(words)) # № 6

print (len([word for word in words if words.count (word) == 1])) # №7

print (words [-10:0]) # №8


task_nine = ''
for w in words:
    task_nine += (w + '\n')
print (task_nine) # №9

print (task_nine[19:31]) # №10

final_list = []
for i in range (len(words)-1):
    final_list.append((words [i], words[i+1]))
print (final_list)
```
